# Scrape using Wayback CDX Server API

```python
import sys
import requests as rq
from bs4 import BeautifulSoup as bs
from time import sleep
from time import time
from random import randint
from warnings import warn
import json
import pandas as pd
```

```python
# Indeed Wayback machine archive urls
#url = 'http://web.archive.org/cdx/search/cdx?url=http://indeed.com/jobs?q=&l=los+angeles&from=20190401&to=20190431&output=json'
url = 'http://web.archive.org/cdx/search/cdx?url=http://indeed.com/jobs?q=&l=los+angeles&output=json'
```

```python
# Indeed Wayback machine archive urls
urls = rq.get(url).text
# parses the JSON from urls.
parse_url = json.loads(urls) 
```

```python
# Extract timstamp and orig cols from urls
url_list = []
for i in range(1,len(parse_url)):
    orig_url = parse_url[i][2]
    tstamp = parse_url[i][1]
    waylink = tstamp+'/'+orig_url
    url_list.append(waylink)
```

```python
final_url_list = []
# Compiles final url pattern.
for url in url_list:
    final_link = 'https://web.archive.org/web/'+url #+'&l=los+angeles'
    final_url_list.append(final_link)
```
